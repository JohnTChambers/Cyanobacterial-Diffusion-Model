```python
import matplotlib.pyplot as plt
import math
import numpy as np
import random
import pylab
import pandas as pd
import random
import math 

from matplotlib.animation import FuncAnimation

# initialize parameters
A = 0.1 #initial chemical concentration per active agent
B = 0
source_D = 8 # source diffusion
D = 0.7 #diffusion coefficient
distance = 0.025 # define speed and distance of bacterial units
delta_t = 0.2 #time inegration constant
delta_x = 0.1 #size of a cell in the simulation
delta_y = 0.1
model_time = 3 #time length model is allowed to run
min_sep = 3
population_low = (2) # define number of bacteria range per string:
population_high = (7)
source_n = (2) # define number of nutrient sources:
n_strings = (15) # define number of bacterial strings:
min_x = -25
max_x = 25
min_y = -25
max_y = 25

# define a bacterium class with 2x attributes; status (sink/source) and coord
class bacterium:
  """ A class that describes individual bacteria """
  def __init__(self, coord, status):
    self.status =  status
    self.coord = coord

# Polymer style inital coordinates
def initial_configuration(initial_min_sep, origin, orientation, string_size):
  """ 
  Populates an array with coordinates 'min_sep' distance away from each other in the direction of the orientation of the array

  :param initial_min_sep: The initial minimal seperation between each individual unit in the array
  :param origin: The x,y origin of the array
  :param orientiation: The direction of the array
  :param string_size: The number of inidivudal units in the array

  :return x: numpy array matching the input characteristics
  """
  x = np.zeros((string_size, 2))
  x[0][0] = origin[0]
  x[0][1] = origin[1]
  for i in range(0,string_size):
      x[i][0] = x[0][0] + math.cos(orientation)*initial_min_sep*i 
      x[i][1] = x[0][1] + math.sin(orientation)*initial_min_sep*i
  return x


def settup_polymer_chains(n_strings, population_low, population_high, xt):
  """ define the origin and orientation(angle) of each polymer chain 
  
  :param n_strings: the number of bacterial polymer strings to simulate
  :param population_low: the minimum number of individual units per polymer
  :param population_hight: the maximum number of individual units per polymer
  :param xt: a numpy array grid representing the current state of the simulation

  :return x: a numpy array grid representing the current state of the simulation
  """

  x = np.zeros((0,2))
  # for each string
  for i in range(0, n_strings):
    # define origin and orientation(angle) of the polymer chain
    origin = np.array([random.randint(-25,25), random.randint(-25,25)])
    orientation = (random.randint(0, 360))
    # define size of string
    string_size = random.randint(population_low, population_high)
    # use initial_configuration function to make initial configs
    xt = initial_configuration(min_sep, origin, orientation, string_size)
    # concetnate the strings together
    x = np.concatenate((x, xt))
  return(x)

def settup_bacterial_sinks(x):
  """
  create a list of all bacteria and create a class instance for each  

  :param x: a numpy array representing the current state of the simulation
  
  :return bact_list: a list of each bacterial individual
  """
  bact_list=[]
  for i in range(0, (len(x))):
    name = ('bact_'+str(i))  
    name = bacterium([(x[i][0]), (x[i][1])],-1)
    bact_list.append(name)
  for i in range(0,source_n):
    name = ('source'+str(i))
    # coordinates
    x_coord, y_coord = (random.randint(-14,-12), random.randint(-14,-12))
    name = bacterium([x_coord, y_coord], +1)
    bact_list.append(name)
    light_pos1 = [x_coord, y_coord]
  return (bact_list)


# diffusion simulation
def conc_simulation2(A,B,D, c, t_array, bact_list):
    """ Initializes and executes the simulation

    :param A: The initial concentration of nutrient per agent
    :param B: The initial concentration of nutrient per source
    :param D: The initial diffusion co-efficient
    :param c: The grid on which the simlulation is too take place 
    :param t_array: An array containing the locations of each bacterial agent
    :param bact_list: A list of all bacterial agents

    :return c: The modified grid containing the state of the simulation following each round of manipulation
    """
    for t_i in range(2,len(t_array)):
        t = t_array[t_i]
        curr_t = t
        # for each bacteria in bacteria_list
        for i in bact_list:
          # introduce movement towards light source
          x_total = light_pos1[0] - i.coord[0]
          y_total = light_pos1[1] - i.coord[1] 
          i.coord[0] = i.coord[0] + (x_total*distance)
          i.coord[1] = i.coord[1] + (y_total*distance)
          # use the class values we created earlier        
          curr_c = (i.status)*(A/(curr_t**0.5))*np.exp(-(((x-i.coord[0])**2)+((y-i.coord[1])**2))/(4*D*curr_t))
          c = np.add(c, curr_c)
        plt.imshow(c, cmap = 'afmhot', interpolation='nearest')
        plt.show()
    return(c)

# set up the simulation
xt = initial_configuration(min_sep, origin, orientation, string_size)
x = settup_polymer_chains(n_strings, population_low, population_high, xt)
bact_list = settup_bacterial_sinks(x)

#initialize grid coordinates
t_array = np.arange(start=0, stop=model_time, step=delta_t)
xvalues = np.arange(start=min_x, stop=max_x, step=delta_x)
yvalues = np.arange(start=min_y, stop=max_y, step=delta_y)
x, y = np.meshgrid(xvalues, yvalues)
c = np.zeros([500,500])

# call the main function
conc_simulation2(A,B,D, c, t_array, bact_list)
```
